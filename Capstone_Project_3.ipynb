#Regression Model

###Method:
Two regression models were trained and evaluated in parallel:

- Linear Regression - used as a baseline predictive model
- Boosted Decision Tree Regression - used to capture more complex, non-linear relationships

Both models used the following predictors:

- Price
- Promotion status
- Seasonality
- Product placement
- Product category information

regressionmodelscreenshot.png

###Linear Wins:

The Linear Regression model achieved the strongest overall performance, demonstrating:

- Lower error rates (MAE, RMSE, and RAE)

- Higher R² values

- More stable and consistent predictions across evaluation metrics

*Linear Regression provided more reliable forecasting performance for this dataset.*

The linear regression pipeline was used to predict Zara product sales, mainly focusing on product placement. After training the model with features like price, promotion, seasonality, and sub-category, a web service input was used to test different store placements such as Aisle, Front of Store, and End-Cap. The pipeline then produced predicted sales in the “Scored Labels” column for each case.

#Hypothesis 1: Product Placement For Sales Growth

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_parquet("data.dataset.parquet")

palette = ["#8A99A4", "#603A5B", "#525671"]

plt.figure(figsize=(8,5))

sns.boxplot(
    x="Product_Position",
    y="Scored Labels",
    data=df,
    hue="Product_Position",
    palette=palette,
    legend=False
)
plt.grid(True, linestyle="--", alpha=0.4)

plt.title("Hypothesis 1: Product Placement vs Sales Growth")
plt.xlabel("Product Position")
plt.ylabel("Predicted Sales")

plt.show()
```

This boxplot compares predicted sales across different product placement positions within the store. The visualization highlights how placement influences sales performance and identifies whether certain store locations generate stronger predicted sales outcomes than standard aisle placements.

#Hypothesis 2: Product Category to Growth

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_parquet("data.dataset.parquet")

palette = ["#8A99A4", "#603A5B", "#525671", "#605046"]

plt.figure(figsize=(10,5))

sns.boxplot(
    x="Sub_Category",
    y="Scored Labels",
    data=df,
    hue="Sub_Category",
    palette=palette,
    legend=False
)
plt.grid(True, linestyle="--", alpha=0.4)

plt.title("Hypothesis 2: Product Category vs Sales Growth")
plt.xlabel("Sub Category")
plt.ylabel("Predicted Sales")
plt.xticks(rotation=45)

plt.show()
```

This boxplot shows how predicted sales vary across different product sub-categories. It helps identify which categories tend to generate higher or more consistent sales based on the model’s predictions.  Showing that some sub-categories have more stable performance while others fluctuate more widely.

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.transform import factor_cmap
import pandas as pd

df = pd.read_parquet("data.dataset evalhyp3.parquet")
df["Promotion"] = df["Promotion"].astype(str)

source = ColumnDataSource(df)

p = figure(
    title="Price vs Sales (Promotion Effect)",
    x_axis_label="Price",
    y_axis_label="Sales",
    width=700,
    height=450
)

p.scatter(
    x="price",
    y="Sales",
    source=source,
    size=6,
    alpha=0.5,
    color=factor_cmap(
        "Promotion",
        palette=["#8A99A4", "#603A5B"],
        factors=["True", "False"]
    )
)

show(p)
```

hyp3scatter.png

This scatter plot compares Sales and Price by Promotion status to evaluate whether promotions significantly influence sales growth. The results show substantial overlap between promoted and non-promoted products, suggesting that promotions do not create a strong or consistent increase in overall sales performance.

Although some promoted products achieved higher sales values, the effect was not consistent across all price levels. This indicates that other factors, including product category, pricing strategy, and demand variability. Overall, the findings suggest that promotions have a limited and context-dependent impact on sales growth rather than acting main driver of sales.

#Key Takeaways

###Placement (Strongest Driver)

- Aisle placements drive steady but lower sales performance.
- End-cap placements show moderate and less consistent sales outcomes.

###Products Driving Sales:
- Sales performance varies across categories without a dominant category.
- Some categories show stable mid-to-high performance across placements.

###Promotional Impact:
- Promotions show limited impact on increasing overall sales levels.
- Promoted and non-promoted products display overlapping sales performance.
- Promotional effects are inconsistent across price levels and categories.
- Promotions are better for driving customer traffic and transaction frequency than increasing average spend per transaction.

###Predictive Modeling:
- Sales relationships appear more linear than highly non-linear.
- Overall model performance suggests additional external factors influence sales.
- Results indicate stronger performance at mid-range price points. Zara may benefit from maintaining competitive pricing in this range to support higher sales performance.

#Business Recommendations

###Better Product Placement:
- Prioritize front-of-store placements for top-selling products to maximize visibility and revenue. Use aisle and end-cap areas for steady-performing items for a balanced store flow.

###Pricing Strategy:
- Focus on mid-range pricing to drive the highest overall sales volume while maintaining select premium pricing for higher-margin transactions.

###Promotional Targeting:
- Use promotions primarily to attract price-sensitive customers and increase transaction frequency rather than expecting higher spend per purchase.
